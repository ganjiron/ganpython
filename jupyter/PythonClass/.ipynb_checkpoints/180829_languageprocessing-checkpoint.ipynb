```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def get_cse_labs(url):
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    content = soup.select_one("#content")
    
    headers = [header.text.strip() for header in content.find_all('th')]
    rows = content.find('tbody').find_all('tr')
    items = [{head: value.get_text().strip() for head, value in 
              zip(headers, row.find_all('td'))} for row in rows]
    
    return items
```

```python
url = 'https://cse.snu.ac.kr/research/labs'
table = get_cse_labs(url)
df = pd.DataFrame(table, columns= table[0].keys())
```
